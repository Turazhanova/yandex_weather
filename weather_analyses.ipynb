```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('weather_month.csv')
df
```

Data cleaning

```python
columns = df.columns
for column in columns:
    df[column] = df[column].apply(lambda x: str(x).split(' ')[0])
    if column != 'Humidity':
        df[column] = pd.to_numeric(df[column].str.replace('−', '-'), errors='coerce')
    else:
        df[column] = pd.to_numeric(df[column].str.replace('%',''), errors='coerce')
df
```

Finding average of day and night climates, coldest day and hottest day using pandas

```python
average_weather = df[['Day-climate', 'Night-climate']].mean()
coldest_day = df[df['Day-climate'] == df['Day-climate'].min()]
hottest_day = df[df['Day-climate'] == df['Day-climate'].max()]
print(average_weather, coldest_day, hottest_day)
```

Visualization using matplotlyb

```python
plt.figure(figsize=(10, 6))

plt.plot(df['Day'], df['Day-climate'], label='Day Climate')
plt.plot(df['Day'], df['Night-climate'], label='Night Climate')
plt.plot(df['Day'], df['Feel-like-climate'], label='Feel-like Climate')

plt.title('Monthly Weather Data')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
